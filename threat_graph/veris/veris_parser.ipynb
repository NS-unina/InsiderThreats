```python
# Write columns
veris.columns
with open('columns.txt', 'w') as out_file:
    out_file.writelines([v + "\n" for v in veris.columns])
```

```python
types = veris.dtypes
with open('types.txt', 'w') as out_file:
    out_file.writelines([str(v) + "\n" for v in types])
```

```python
def last_year(veris):
    v2020= veris['plus.timeline.notification.year'] == 2020
    return veris[v2020]
```

```python
last_veris = last_year(veris)
print(last_veris.shape[0])
last_veris.to_csv('2020.csv')
```

Filter out empty values

```python
columns = last_veris.columns 
last_veris_copy = last_veris.copy()
def is_empty(curr_col, type_str):
    empty = True
    for c in curr_col:
        if type_str == 'bool':
            if c != False:
                empty = False
        elif type_str == 'object':
            if not pd.isnull(c):
                empty = False
    return empty
            

for c in columns:
    d = last_veris_copy[c]
    if is_empty(d, d.dtypes.name):
        print("{} is empty, remove it.".format(c))
        last_veris_copy.drop(c, inplace=True, axis=1)
    

```

```python
last_veris_copy.to_csv('2020_filtered.csv')
```

```python
print(last_veris.shape[1])
print(last_veris_copy.shape[1])
```

```python

def row2str(row, columns):
    s = str(row['incident_id']) + "\n"

    for c in columns:
        d = row[c]
        
        if d != False and not pd.isnull(d):
            s += c + " = " + str(d) + "\n"

    s += "=====================\n\n"
    return s

last_veris_copy.reset_index()
columns = last_veris_copy.columns
row = last_veris_copy.iloc(0)
#print_row(row[0], columns)
with open('incidents.txt', 'w') as f: 
    for index, row in last_veris_copy.iterrows():
        s = row2str(row, columns)
        f.write(s)

```
